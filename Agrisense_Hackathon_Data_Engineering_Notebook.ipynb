```python
import pandas as pd
from datetime import datetime

# 1. Load the input CSVs
zones = pd.read_csv('/content/formatted_plant_hardiness_zones_ohio.csv')         # zipcode, zone, trange, zonetitle
regions = pd.read_csv('/content/formatted_ohio_zip_city_region.csv')                       # zipcode, city, region
incentives = pd.read_csv('/content/drive/MyDrive/AgriSense/government_incentives.csv')                   # zipcode, incentive_score
corn_windows = pd.read_csv('/content/sample_data/formatted_corn_planting_windows.csv')       # zone, region, start, end, latest
soy_windows  = pd.read_csv('/content/formatted_soybean_planting_windows.csv')    # zone, region, start, end, latest
climate = pd.read_csv('/content/ohio_climate_normals_with_growing_season.csv')   # Month, Avg_High_Temp_F, Avg_Low_Temp_F, Avg_Precip_Inches, ...

# 2. Compute statewide precipitation metrics (average annual precip & median)
avg_annual_precip = climate['Avg_Precip_Inches'].sum()
avg_monthly_precip = climate['Avg_Precip_Inches'].mean()  # average monthly
median_precip = climate['Avg_Precip_Inches'].median()

# 3. Prepare planting-window functions
def calculate_window_days(row):
    """
    Compute the length of a planting window in days given start and end strings (MM-DD).
    Uses a dummy year (2024).
    """
    start_dt = datetime.strptime(row['Recommended_Planting_Start'] + '-2024', '%m-%d-%Y')
    end_dt   = datetime.strptime(row['Recommended_Planting_End'] + '-2024', '%m-%d-%Y')
    # If the end date is earlier in the year (shouldn't happen here), adjust year
    if end_dt < start_dt:
        end_dt = datetime.strptime(row['Recommended_Planting_End'] + '-2025', '%m-%d-%Y')
    return (end_dt - start_dt).days + 1

# Apply to corn and soybean windows
corn_windows['corn_window_days'] = corn_windows.apply(calculate_window_days, axis=1)
soy_windows['soy_window_days']   = soy_windows.apply(calculate_window_days, axis=1)

# 4. Merge data by ZIP and zone
df = zones[['zipcode', 'zone']].merge(regions[['zipcode', 'region']], on='zipcode', how='left')
df = df.merge(incentives[['zipcode', 'incentive_score']], on='zipcode', how='left')
# Join corn & soybean windows on zone (drop Region column to avoid duplicate)
df = df.merge(corn_windows[['Zone', 'corn_window_days']], left_on='zone', right_on='Zone', how='left')
df = df.merge(soy_windows[['Zone', 'soy_window_days']], left_on='zone', right_on='Zone', how='left')
df.drop(columns=['Zone_x', 'Zone_y'], inplace=True)

# 5. Compute sustainability scores
def compute_scores(row):
    scores = {}
    # Water resilience: compare against statewide average monthly precipitation
    water_score = 0
    if avg_monthly_precip > 0:
        water_score = 2 if avg_monthly_precip >= median_precip else 1
    # Add 1 point if incentive score is high (>= 0.7)
    if row['incentive_score'] >= 0.70:
        water_score += 1

    # Climate adaptability: longer planting windows for both crops
    climate_score = 0
    if row['corn_window_days'] >= 20 and row['soy_window_days'] >= 20:
        climate_score = 2
    else:
        climate_score = 1

    # Crop system resilience: ability to plant both crops (exists for all zones in this dataset)
    crop_resilience = 2

    # Soil health potential: zone-based scoring
    soil_score = 1  # default
    if row['zone'] in ['6a', '6b']:
        soil_score = 2
    # Add 1 point for high incentive score (>= 0.7) as a proxy for adoption of soil health practices
    if row['incentive_score'] >= 0.70:
        soil_score += 1

    scores['water_resilience_score'] = water_score
    scores['climate_adaptability_score'] = climate_score
    scores['crop_system_resilience_score'] = crop_resilience
    scores['soil_health_score'] = soil_score
    scores['sustainability_score'] = water_score + climate_score + crop_resilience + soil_score
    return pd.Series(scores)

# Fill missing incentive scores with 0 before computing
df['incentive_score'] = df['incentive_score'].fillna(0)
score_cols = df.apply(compute_scores, axis=1)
df = pd.concat([df, score_cols], axis=1)

# 6. Generate sustainability tags based on scores
def tag_row(row):
    tags = []
    if row['water_resilience_score'] >= 3:
        tags.append('high-water-resilience')
    elif row['water_resilience_score'] == 2:
        tags.append('moderate-water-resilience')
    else:
        tags.append('low-water-resilience')
    if row['climate_adaptability_score'] == 2:
        tags.append('long-planting-window')
    else:
        tags.append('short-planting-window')
    if row['crop_system_resilience_score'] == 2:
        tags.append('diverse-cropping')
    if row['soil_health_score'] >= 3:
        tags.append('loamy-zone')
    else:
        tags.append('cold-zone')
    return ';'.join(tags)

df['sustainability_tags'] = df.apply(tag_row, axis=1)

# 7. Order columns and save to CSV
df_final = df[['zipcode', 'region', 'zone', 'corn_window_days', 'soy_window_days',
               'incentive_score', 'water_resilience_score', 'climate_adaptability_score',
               'crop_system_resilience_score', 'soil_health_score', 'sustainability_score',
               'sustainability_tags']].copy()

df_final.to_csv('sustainability_scores.csv', index=False)
print("Output written to sustainability_scores.csv")
```

## Group data and calculate mean incentive score

### Subtask:
Group the DataFrame by 'corn_window_days', 'soy_window_days', and 'sustainability_score' and calculate the mean 'incentive_score' for each group, excluding the 0.00 values.


**Reasoning**:
Filter the DataFrame to exclude rows with 'incentive_score' of 0.00 and then group by the specified columns to calculate the mean 'incentive_score' for each group.



```python
# Filter out rows where incentive_score is 0.00
filtered_df_for_mean = df[df['incentive_score'] != 0.00].copy()

# Group by 'corn_window_days', 'soy_window_days', and 'sustainability_score' and calculate the mean 'incentive_score'
grouped_incentive_means = filtered_df_for_mean.groupby(['corn_window_days', 'soy_window_days', 'sustainability_score'])['incentive_score'].mean()

display(grouped_incentive_means)
```

## Impute 0.00 incentive scores

### Subtask:
Iterate through the DataFrame and replace the 0.00 'incentive_score' values with the calculated group means.


**Reasoning**:
Iterate through the DataFrame and replace the 0.00 'incentive_score' values with the calculated group means.



```python
for index, row in df.iterrows():
    if row['incentive_score'] == 0.00:
        try:
            mean_incentive = grouped_incentive_means.loc[(row['corn_window_days'], row['soy_window_days'], row['sustainability_score'])]
            df.loc[index, 'incentive_score'] = mean_incentive
        except KeyError:
            # Handle cases where a group mean doesn't exist for this combination
            pass

display(df.head())
```

## Handle groups with no non-zero incentive scores

### Subtask:
For groups where all 'incentive_score' values were 0.00, decide on an imputation strategy (e.g., impute with the overall mean incentive score or a different value).


**Reasoning**:
Identify rows with remaining 0.00 'incentive_score', calculate the overall mean incentive score from the filtered data, and impute the remaining 0.00 values with this overall mean. Then display the head of the updated DataFrame.



```python
# Identify rows where 'incentive_score' is still 0.00
remaining_zero_incentive_rows = df[df['incentive_score'] == 0.00]

# Calculate the overall mean of 'incentive_score' from the original filtered_df_for_mean
overall_mean_incentive = filtered_df_for_mean['incentive_score'].mean()

# Replace the remaining 0.00 'incentive_score' values in df with this overall mean
df['incentive_score'] = df['incentive_score'].replace(0.00, overall_mean_incentive)

# Display the head of the updated DataFrame
display(df.head())
```

```python
# Check if there are any remaining rows with 0.00 'incentive_score'
remaining_zeros_check = df[df['incentive_score'] == 0.00]

if remaining_zeros_check.empty:
    print("Imputation successful: No remaining 0.00 'incentive_score' values found.")
else:
    print("Imputation may not be fully successful. Remaining 0.00 'incentive_score' values found:")
    display(remaining_zeros_check)
```
